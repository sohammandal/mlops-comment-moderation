### Test Data

```python
test_labels_df = pd.read_csv('/Users/sohammandal/Class/MLOps/jigsaw-toxic-comment-classification-challenge/test_labels.csv')
test_labels_df.head()
print(test_labels_df.shape)
```

```python
toxicity_cols = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']

# Step 1: Filter out rows where all labels are -1 (i.e., unscored)
test_labels_df = test_labels_df[~(test_labels_df[toxicity_cols] == -1).all(axis=1)].copy()

# Step 2: Create 'moderation_label' — set to 1 if any individual label is 1, else 0
test_labels_df['moderation_label'] = (test_labels_df[toxicity_cols] == 1).any(axis=1).astype(int)

test_labels_df.head()
```

```python
print(test_labels_df.shape)
test_labels_df.toxic.value_counts()    
```

```python
test_df_combined = test_df.merge(test_labels_df, on='id', how='inner')
test_df_combined
```
